```python
from datosgobmx import client
import pandas as pd
```

```python
import datosgobmx.client as api
```

```python
api.api("endpoint", {"page": 1})
```

```python
client
```

```python
parametrosRequest=client.makeCall('sinaica-parametros')

validParams=[]
for v in parametrosRequest['results']:
    aux=pd.DataFrame.from_dict(v,orient='index').T
    validParams.append(aux)
validParams=pd.concat(validParams,ignore_index=True)
validParams['date-insert']=pd.to_datetime(validParams['date-insert'])

validParams
```

# Estaciones

```python
estaciones=client.makeCall('sinaica-estaciones', {'pageSize':200})

validEstaciones=[]
for v in estaciones['results']:
    aux=pd.DataFrame.from_dict(v,orient='index').T
    validEstaciones.append(aux)
validEstaciones=pd.concat(validEstaciones,ignore_index=True)


validEstaciones
```

```python
import numpy as np
```

```python
pip install folium
```

```python
import folium
```

```python
centro_lat, centro_lon=22.396092,-101.731430
folium_map=folium.Map(location=[centro_lat,centro_lon],zoom_start=5)

for i, row in validEstaciones.iterrows():
    popupTtext=f"<b> Nombre:</b>{row.nombre}<br><b> Latitud:</b> {row.lat:.5f}<br><b>Longitud: </b>{row.long:.5f}<br>"
    folium.CircleMarker(location=[row.lat,row.long],radius=5,tooltip=popupTtext,fill=True,fill_capacity=.4).add_to(folium_map)
    
folium_map
```

```python
cities=client.makeCall('sinaica', {'pageSize':3000})

citiesDF=[]
for v in cities ['results']:
    aux=pd.DataFrame.from_dict(v,orient='index').T
    
    citiesDF.append(aux)
citiesDF=pd.concat(citiesDF,ignore_index=True)


citiesDF
```

# O3

```python
paramO3=client.makeCall('sinaica', {'pageSize':500, 'parametro':'O3', "date":"2018-01-30"})
paramO3
```

```python
paramO3_df=[]
for p in paramO3 ['results']:
    aux=pd.DataFrame.from_dict(p,orient='index').T
    paramO3_df.append(aux)
    
paramO3_df=pd.concat(paramO3_df,ignore_index=True)
paramO3_df
```

```python
O3_df = pd.merge(validEstaciones,paramO3_df,left_on='id',right_on='estacionesid')
O3_df
```

```python
O3_df.to_csv ('O3.csv', index = None)
```

```python
centro_lat, centro_lon=22.396092,-101.731430
folium_map_1=folium.Map(location=[centro_lat,centro_lon],zoom_start=5)

for i, row in O3_df.iterrows():
    popupTtext=f"<b> Nombre:</b>{row.nombre}<br><b> Ciudad:</b>{row.city}<br><b> Valor:</b>{row.valororig}<br>"
    folium.CircleMarker(location=[row.lat,row.long],radius=5,tooltip=popupTtext,fill=True,fill_capacity=.4).add_to(folium_map_1)
    
folium_map_1
```
